```python
import nltk
nltk.download('punkt_tab')
```

```python
# Importare le librerie necessarie
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from transformers import pipeline

# Caricare il file
with open("prova_testo.txt", "r", encoding="utf-8") as file:
    text = file.read()

print("Testo caricato:")
print(text)  # Mostra i primi 500 caratteri
```

```python
# Utilizzare la libreria NLTK per la segmentazione in frasi
sentences = sent_tokenize(text)

print("Frasi segmentate:")
print(sentences)
```

```python
# Tokenizzare le frasi
tokens = [word_tokenize(sentence) for sentence in sentences]

print("Esempio di tokenizzazione:")
print(tokens)  # Mostra i token delle prime 5 frasi
```

```python
# Utilizzare Spacy per il POS tagging
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

print("Esempio di POS tagging:")
for token in doc:
    print(f"{token.text}: {token.pos_}")  # Mostra i primi 10 token con POS
```

```python
# Estrazione di parole chiave usando Spacy
keywords = [chunk.text for chunk in doc.noun_chunks]

print("Parole chiave estratte:")
print(keywords)  # Mostra le prime 10 parole chiave
```

```python
# Named Entity Recognition con Spacy
entities = [(ent.text, ent.label_) for ent in doc.ents]

print("Entità riconosciute:")
print(entities)  # Mostra le prime 10 entità
```

```python
# Analizzare le relazioni sintattiche
print("Relazioni sintattiche:")
for token in doc:
    print(f"{token.text} -> {token.dep_} -> {token.head.text}")
```

```python
# Utilizzare NER per disambiguare entità (Spacy aiuta in questo processo)
for ent in entities:
    print(f"Entità: {ent[0]} - Tipo: {ent[1]}")
```

```python
# Relazioni semplici tra entità usando dipendenze
relations = []
for token in doc:
    if token.dep_ in ("nsubj", "dobj"):
        relations.append((token.head.text, token.text))

print("Relazioni trovate:")
print(relations)  # Mostra le prime 10 relazioni
```

```python
# Individuazione di eventi
events = [token for token in doc if token.pos_ == "VERB"]

print("Eventi trovati (verbi):")
print(events)  # Mostra i primi 10 eventi
```

```python
# Summarization usando Hugging Face Transformers
summarizer = pipeline("summarization", model="allenai/led-large-16384")

summary = summarizer(text, max_length=150, min_length=50, do_sample=True, temperature=0.7)

print("Riassunto:")
print(summary[0]["summary_text"])
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(text, summary[0]["summary_text"])

print("Metriche ROUGE:")
print("ROUGE-1:", scores['rouge1'].fmeasure)
print("ROUGE-2:", scores['rouge2'].fmeasure)
print("ROUGE-L:", scores['rougeL'].fmeasure)
```

```python
from nltk.translate.bleu_score import sentence_bleu

original_sentences = text.split(". ")
summary_sentences = summary[0]["summary_text"].split(". ")

bleu_score = sentence_bleu([original_sentences], summary_sentences)
print("BLEU Score:", bleu_score)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Carica il modello di embedding
model = SentenceTransformer('all-MiniLM-L6-v2')

# Calcola gli embedding per il testo originale e il riassunto
original_embedding = model.encode(text)
summary_embedding = model.encode(summary[0]["summary_text"])

# Calcola la similarità coseno
similarity = cosine_similarity([original_embedding], [summary_embedding])

print("Cosine Similarity:", similarity[0][0])
```
